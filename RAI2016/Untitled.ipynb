```python
from datascience import *
from tqdm import *
import pandas as pd
import numpy as np
import re
```

```python
with open ('../Scrape-Oracc/Input/amarna.txt') as f:
    amarna = f.read().splitlines()
amarna_table = Table(['id_text','text_name','l_no','text'])
for text in tqdm(amarna):
    text = text.replace('/', '_')
    onetext = Table.read_table('../Scrape-Oracc/output/' + text + '.txt', sep=',')
    amarna_table = amarna_table.append(onetext)
```

```python
amarna_table = amarna_table.where('text', are.not_equal_to('nan'))
amarna_text = amarna_table['text']
amarna_text = [line.lower() for line in amarna_text]
amarna_table.append_column('text', amarna_text)
amarna_table
```

```python
amarna_letters = amarna_table.select(['text_name', 'text']).group('text_name', collect=' '.join)
amarna_letters
```

```python
amarna_pn = 'Source,Target,Textid\n'
for index, item in enumerate(amarna_letters['text join']):
    pn_set = set(amarna_letters['text join'][index].split())
    pn_unique_list = list(pn_set)
    #textid = amarna_letters['text_name'][index].replace('aemw/amarna/', "")
    textid = amarna_letters['text_name'][index]
    textid = re.sub(r'.*(\[.*\])', r'\1', textid)
    for i in range(len(pn_unique_list)-1):
        for j in range(i+1,len(pn_unique_list)):
            amarna_pn = amarna_pn + pn_unique_list[i] + ',' + pn_unique_list[j] + ',' + textid +'\n'
amarna_pn
```

```python
with open('../Scrape-Oracc/output/amarna_edges.csv', mode='w') as writeFile:
    writeFile.write(amarna_pn)
```

```python
node_set = []
for index, item in enumerate(amarna_letters['text join']):
    pn_set = amarna_letters['text join'][index].split()
    for i in range(len(pn_set)):
        node_set.append(pn_set[i])
node_set = list(set(node_set))
amarna_nodes = 'id\n'
for pn in node_set:
    amarna_nodes = amarna_nodes + pn + '\n'
```

```python
with open('../Scrape-Oracc/output/amarna_nodes.csv', mode='w') as writeFile:
    writeFile.write(amarna_nodes)
```
